Благодаря атрибуту columns мы получили массив с наименованием наших колонок. Здесь есть track_id - ID трека, track_name - название трека, track_artist - исполнителя, track_popularity - популярность,track_album_id - ID альбома, track_album_name - название альбома, track_album_release_date - дату релиза,

Благодаря атрибуту nunique мы получили все уникальные значения

```python
spotify_songs.nunique()
```

С помощью атрибута sort_values мы отсортировали значения в порядке возрастания по столбцу track_popularity

```python
spotify_songs.sort_values(by='track_popularity')
```

С помощью атрибута drop мы удалили столбец track_id

```python
spotify_songs.drop(columns=["track_id"])
```

С помощью атрибута replace мы меняем все dance pop в столбце playlist_subgenre на pop dance

```python
spotify_songs['playlist_subgenre']=spotify_songs['playlist_subgenre'].replace('dance pop', 'pop dance')
```

С помощью атрибута drop_duplicates мы удалили дубликаты

```python
spotify_songs.drop_duplicates()
```

С помощью атрибута info мы получили общую информацию о датасете, включая количество непустых значений, типы данных каждого столбца, общее количество столбцов и количество использованной памяти. Мы видим, что всего 23 столбца, все 32833 данных каждого столбца имеют тип данных not null, количество использованной памяти = 5.8 + MB

```python
spotify_songs.info()
```
